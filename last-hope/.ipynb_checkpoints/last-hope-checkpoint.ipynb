**I continue to follow the guide for help on running this AER simulation and tailoring it with the data we have**
https://qiskit.org/textbook/ch-appendix/qiskit.html#Classical-registers-and-the-Aer-simulator

```python
# Load the QASM file as a string
with open('quantum_artifact.qasm', 'r') as file:
    qasm_string = file.read()

# Create a quantum circuit object from the QASM string
qreg = QuantumRegister(2)
creg = ClassicalRegister(2)
qc = QuantumCircuit(qreg, creg)
qc = QuantumCircuit.from_qasm_str(qasm_string)

# Run the circuit on a local simulator
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend)
result = job.result()

# Print the measurement counts
count = result.get_counts(qc)
print(count)
```
